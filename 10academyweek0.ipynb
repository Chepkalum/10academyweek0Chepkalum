```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
# Load the data
df = pd.read_csv("rating.csv")
```

```python
# Top 10 websites with the largest count of news articles
top_10_articles = df['source_name'].value_counts().head(10)
print("Top 10 Websites with Largest Count of News Articles:")
print(top_10_articles)
```

```python
# Bottom 10 websites with the smallest count of news articles
bottom_10_articles = df['source_name'].value_counts().tail(10)
print("Bottom 10 Websites with Smallest Count of News Articles:")
print(bottom_10_articles)
```

```python
# Top 10 websites with the highest numbers of visitors traffic
top_10_traffic = df.groupby('source_name')['url'].count().nlargest(10)
print("Top 10 Websites with Highest Numbers of Visitors Traffic:")
print(top_10_traffic)
```

```python
# Bottom 10 websites with the lowest numbers of visitors traffic
bottom_10_traffic = df.groupby('source_name')['url'].count().nsmallest(10)
print("Bottom 10 Websites with Lowest Numbers of Visitors Traffic:")
print(bottom_10_traffic)
```

```python
df['domain'] = df['source_name'].str.split('.').str[-1]
# Mapping of Domain Extensions to Countries
domain_country_mapping = {
    'com': 'United States',
    'uk': 'United Kingdom',
    'ca': 'Canada',
    'af': 'Africa',
    'ch':'China',
    'rus':'Russia'
   
}

# Infer Countries from Domain Names
def infer_country(domain):
    extension = domain.split('.')[-1]
    return domain_country_mapping.get(extension, 'Unknown')

df['country'] = df['domain'].apply(infer_country)

# Count the Occurrences of Media Organizations for Each Country
country_counts = df['country'].value_counts()

# Display the Top 10 Countries
top_10_countries = country_counts.head(10)
print("Countries with the Highest Number of News Media Organizations:")
print(top_10_countries)
```

```python
# Countries with many articles written about them
countries_articles = df['source_name'].value_counts()
print("Countries with Many Articles Written About Them:")
print(countries_articles)
```

```python
# Websites that reported about specific countries
specific_countries = df[df['Country'].isin(['Africa', 'US', 'China', 'EU', 'Russia', 'Ukraine', 'Middle East'])]
specific_countries_count = specific_countries['source_name'].value_counts()
print("Websites that Reported About Africa, US, China, EU, Russia, Ukraine, Middle East:")
print(specific_countries_count)
```

```python
# Sentiment distribution
sentiment_distribution = df['title_sentiment'].value_counts()
print("Sentiment Distribution:")
print(sentiment_distribution)
```

```python
articles = df['source_name'].value_counts().head(1000)

# Extract the website names and their corresponding counts
website_names = articles.index.tolist()
article_counts = articles.values.tolist()

# Create a scatter plot
plt.figure(figsize=(12, 6))
sns.scatterplot(x=website_names, y=article_counts, color='blue')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Set plot title and labels
plt.title('Scatter plot for  Websites')
plt.xlabel('Source Name')
plt.ylabel('Number of Articles')

# Adjust layout to prevent clipping of labels
plt.tight_layout()


```
